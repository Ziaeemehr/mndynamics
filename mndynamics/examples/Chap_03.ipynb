```python
obj = HH_GATING_VARIABLES({"i_ext": 10.0})
data = obj.simulate(tspan=np.arange(0, 50, 0.01))
data.keys()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 4, figsize=(15, 3))

t = data['t']
v = data['v']
m = data['m']
h = data['h']
n = data['n']

ax[0].plot(t, v)
ax[1].plot(t, m)
ax[2].plot(t, h)
ax[3].plot(t, n)
labels = ['v', 'm', 'h', 'n']
for i in range(4):
    ax[i].margins(x=0)
    ax[i].set_ylabel(labels[i])
plt.tight_layout()
```
