```python
import numpy as np 
import matplotlib.pyplot as plt
from mndynamics.models.py.ch05 import WB_GATING_VARIABLES
from mndynamics.models.py.ch05 import RTM_GATING_VARIABLES
from mndynamics.models.py.ch05 import Erisir_GATING_VARIABLES
```

```python
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['k', 'g', 'b', 'y'])
plt.rcParams['figure.figsize'] = (10, 8)
plt.rcParams['lines.linewidth'] = 2.0
plt.rcParams['font.size'] = 16
```

```python
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(7, 7))

rtm = RTM_GATING_VARIABLES()
rtm.plot(ax, color="r", ls="-", label="RTM")
wb = WB_GATING_VARIABLES()
wb.plot(ax, color="b", ls=":", label="WB")
er = Erisir_GATING_VARIABLES()
er.plot(ax, color="k", ls="--", label="Erisir")
```
