```python
import numpy as np 
import matplotlib.pyplot as plt
from mndynamics.models.py.WB_Base import WB
from mndynamics.models.py.RTM_Base import RTM 
from mndynamics.models.py.Erisir_Base import Erisir
from mndynamics.models.py.Erisir_Base import ErisirM
from mndynamics.models.py.ch05 import WB_GATING_VARIABLES
from mndynamics.models.py.ch05 import RTM_GATING_VARIABLES
from mndynamics.models.py.ch05 import Erisir_GATING_VARIABLES
```

```python
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['k', 'g', 'b', 'y'])
plt.rcParams['figure.figsize'] = (10, 8)
plt.rcParams['lines.linewidth'] = 2.0
plt.rcParams['font.size'] = 16
```

```python
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(7, 7))

rtm = RTM_GATING_VARIABLES()
rtm.plot(ax, color="r", ls="-", label="RTM")
wb = WB_GATING_VARIABLES()
wb.plot(ax, color="b", ls=":", label="WB")
er = Erisir_GATING_VARIABLES()
er.plot(ax, color="k", ls="--", label="Erisir")
```

The functions $x_{\infty}$ and $\tau_x$ for the RTM neuron (red and solid),
WB neuron (blue, dash-dots), and Erisir neuron (black, dashes).

#### RTM Voltage Trace

```python
obj = RTM()
obj()
```

```python
tspan = np.arange(0, 100, 0.01)
data = obj.simulate(tspan)
data.keys()
```

```python
t = data["t"]
v = data["v"]
m = data["m"]
h = data["h"]
n = data["n"]


fig, ax = plt.subplots(2, figsize=(7, 6), sharex=True)
ax[0].plot(t, v, lw=2, c="k")

ax[1].plot(t, m, lw=2, label="m", c="b", alpha=0.6)
ax[1].plot(t, h, lw=2, label="h", c="g", alpha=0.6)
ax[1].plot(t, n, lw=2, label="n", c="r", alpha=0.6)


ax[0].set_xlim(min(t), max(t))
ax[0].set_ylim(-100, 50)
ax[1].set_xlabel("time [ms]")
ax[0].set_ylabel("v [mV]")
ax[0].set_yticks(range(-100, 100, 50))
ax[1].legend()
plt.tight_layout()
```
