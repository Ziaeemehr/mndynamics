```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.size'] = 16
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['figure.figsize'] = [7, 5]
```

```python
# [WB_NETWORK_WITH_GJ]
from mndynamics.models.py.WB_Base import WB_Net_GJ

par = {'adj': np.array([[0, 1], [1, 0]]),
       'i_ext': [1.0, 0.0],
       'v0': [-63.0, -63.0],}

obj = WB_Net_GJ(par)
obj()
```

```python
data = obj.simulate()
t = data['t']
v = data['v']
```

```python
fig, ax = plt.subplots(2, figsize=(7, 5), sharex=True)
ax[0].plot(t, v[:,0], lw=2, c="k")
ax[1].plot(t, v[:,1], lw=2, c="k")

for i in range(2):
    ax[i].set_xlim(100, max(t))
    ax[i].set_xlabel("time [ms]")
ax[0].set_yticks(range(-100, 100, 50))
ax[0].set_ylabel(r"$v_1$ [mV]")
ax[1].set_ylabel(r"$v_2$ [mV]")
ax[0].set_ylim(-100, 50)
ax[1].set_ylim(-64, -62)
plt.tight_layout()
```
