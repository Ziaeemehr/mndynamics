```python
import os
import warnings
import numpy as np 
import matplotlib.pyplot as plt
from mndynamics.models.py.Inapik_Base import Inapik

warnings.filterwarnings("ignore")

# if not os.path.exists('output'):
#     os.makedirs('output')
```

```python
plt.rcParams['font.size'] = 16
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['figure.figsize'] = [7, 5]
```

```python
obj = Inapik()
obj()
```

```python
data = obj.simulate()
data.keys()
```

```python
t = data["t"]
v = data["v"]

fig, ax = plt.subplots(1, figsize=(8, 3.5))
ax.plot(t, v, lw=2, c="k")


ax.set_xlim(min(t), max(t))
ax.set_ylim(-100, 50)
ax.set_xlabel("time [ms]")
ax.set_ylabel("v [mV]")
ax.set_yticks(range(-100, 100, 50))
plt.tight_layout()
```

```python
obj = Inapik({"i_ext": 7.0, "t_end":1500})
data = obj.simulate()
t = data["t"]
v = data["v"]
n = data["n"]
n_slow = data["n_slow"]
```

```python
# 3D plot (v, n, n_slow)
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot(v, n, n_slow, lw=1, c="k")
ax.set_xlabel("v [mV]")
ax.set_ylabel("n")
ax.set_zlabel("n slow")
ax.set_xlim(-100, 50)
ax.set_yticks([0, 0.3, 0.6])
ax.set_xticks([-100, -50, 0, 50])
plt.tight_layout()
```
